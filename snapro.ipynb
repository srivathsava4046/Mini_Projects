# IMPORT LIBRARIES

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

# LOAD DATASET

```python
df = pd.read_csv('/content/student_data_extended.csv')
```

```python
df.head()
```

```python
df.tail()
```

# REDUCING DATASET

```python
# Set the desired size of the reduced dataset
desired_size = 100

# Check if the current dataset size is greater than the desired size
if len(df) > desired_size:
    # Randomly sample 100 rows from the DataFrame
    reduced_df = df.sample(n=desired_size, random_state=42)  # Set a random state for reproducibility
else:
    reduced_df = df  # If the dataset size is already smaller than 100, keep the original dataset

# Now reduced_df contains the reduced dataset with 100 rows
# You can use reduced_df for further analysis in your Python notebook
```

# REDUCED DATASET

```python
reduced_df.head()
```

# CREATE A EMPTY GRAPH

```python
# Create an empty graph
G = nx.Graph()
```

# ADD NODES AND EDGES

```python
# Add nodes (clubs) and edges (connections)
for index, row in reduced_df.iterrows():
    club = row['ExtracurricularActivities']
    skill = row['GPA']
    G.add_edge(club, skill)
```

# VISUALIZE THE NETWORK

```python
# Visualize the network
pos = nx.spring_layout(G)  # positions for all nodes
nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)
plt.title("Student Club Network")
plt.show()
```

# CALCULATE NETWORK METRICS

```python
# Calculate basic network metrics
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
clustering_coefficient = nx.clustering(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
```

```python
print("Degree Centrality:")
print(degree_centrality)
print("\nBetweenness Centrality:")
print(betweenness_centrality)
print("\nClustering Coefficient:")
print(clustering_coefficient)
print("\nCloseness Centrality:")
print(closeness_centrality)
print("\nEigen vector Centrality:")
print(eigenvector_centrality)
```

# IDENTIFY CENTRAL CLUB

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=[3000 * v for v in degree_centrality.values()], cmap=plt.cm.Blues, node_color=list(degree_centrality.values()), font_size=10)
plt.title("Club Network with Degree Centrality")
plt.show()
```

```python
# Print central clubs based on degree centrality
central_clubs = sorted(degree_centrality, key=degree_centrality.get, reverse=True)[:5]
print("Central Clubs based on Degree Centrality:")
print(central_clubs)
```
